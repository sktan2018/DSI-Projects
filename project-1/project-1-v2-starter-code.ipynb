## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
#Solution 1
#Since all statistics are already captured in players dictionary,
#we don't need data from pokedex dictionary. Hence I modified the 
#function to not take in the pokedex dictionary 
#Assumption made: player_power sums up only values of the
#'attack' and 'defense' base stats

def compute_power1(players, player_id):
    player = players[player_id]
    player_power =0
    for i in player['player_pokemon'].keys():
        player_power += (player['player_pokemon'][i]['attack'] +
        player['player_pokemon'][i]['defense'] )
       
        print('[',player['player_name'],"]'s power is [",player_power,"]")
    return player_power                                   
````

## 9. Write a function to generate the full pokedex

---

Write a function that recreates the pokedex you made before, but with the data read in from the full pokemon file. The `PokedexNumber` should be used as the `pokemon_id` key values for the dictionary of pokemon.

Your function should:

1. Take the parsed pokedex information you created above as an argument.
2. Return a dictionary in the same format as your original pokedex you created before containing the information from the parsed full pokedex file.

To test the function, print out the pokemon with id = 100.

````python
#This function creates the pokedex dictionary with unique 
#pokemon_id key. Since PokedexNumber
#(from clean dictionary) created from the csv file
#is ot unique, ignore PokedexNumber and regenerate
#pokeIdx as unique key 
def create_pokedex (pokedex_info):
    pokedexfull = {} #iniatilise the pokedex
    
    key_row = pokedex_info[0] #the first row of pokedex contains 
                                #the features definition
    data = pokedex_info[1:]
    pokeIdx = 0 
    for row in data :
        pokeIdx += 1
        pokedexfull[pokeIdx] = {}
        for i, col in enumerate(row[1:]): #Ignore the pokedexNumber
            pokedexfull[pokeIdx][key_row[i+1]]= col #Replace the original 
                                        #PokedexNumber with unique values

    return pokedexfull
````

````python
#Given a pokedex and the input criteria (filter dictionary), 
# this function returns all pokedex that satisfy all 
# conditions of each of the input criteria.
#If the input criterion is a float, condition equates to
#true if the corresponding pokedex value >= input criterion.
#If the input criterion, is a string (except for empty string,
# condition equates to True if there is exact match.
#It the input criterion is an empty string, skip.

def filtered_pokedex(pokedex, filter):
    matched_pokedex = []
    
    for i in pokedex.keys(): 
        is_matched = False
         
        for j in filter.keys():
            
            if type(filter[j]) is float or type(filter[j]) is int:  
                if (pokedex[i][j])>=filter[j]:
                    is_matched = True 
                else :
                    is_matched = False
            elif filter[j].strip() != '':
                if (pokedex[i][j])== filter[j]:
                    is_matched = True

                else :
                    is_matched = False

            if not is_matched :
                break
        if is_matched:
            matched_pokedex.append(pokedex[i])
    return matched_pokedex
````

### 9.1

What is the population mean and standard deviation of the "Total" attribute for all characters in the Pokedex?



````python
#This function returns a pokedexdata dictionary with
#keys using the each of the features set
def getPokedexdata(pokedex_info):
    pokedexdata = {}
    header = pokedex_info[0]
    data = pokedex_info[1:]
    
    for index, column_name in enumerate(header):
        pokedexdata[column_name] = []
        for row in data :
            pokedexdata[column_name].append(row[index])
            
    return pokedexdata
````
